```python
#Now i have cleaned up the data, Next step for me is to Identify the outliers if any and remove them
#I am checking some 4-5 columns in one go and plotting box plots to check outliers in them
```

```python
plt.figure(figsize=(14, 8))
for i, col in enumerate(['loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'term', 'int_rate', 'installment']):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(x=dfLoan[col])
    plt.title(f'Box plot for {col}')
plt.show()
```

```python
# As I see there are many outliers in loan_amount, funded-amnt etc, removing them
```

```python
for col in ['loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'term', 'int_rate', 'installment']:        
        Q1 = dfLoan[col].quantile(0.25)
        Q3 = dfLoan[col].quantile(0.75) 
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        dfLoan = dfLoan[(dfLoan[col] >= lower_bound) & (dfLoan[col] <= upper_bound)]
```
